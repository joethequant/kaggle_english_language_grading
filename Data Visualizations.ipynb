```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in dataset
train = pd.read_csv("train.csv")
```

```python
# Inspect output
train.head()
```

```python
# Create histograms
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(7,7))
i = 1
for col in ["cohesion", "syntax", "vocabulary", "phraseology", "grammar", "conventions"]:
    plt.subplot(3,2, i)
    plt.hist(train[col], bins=9)
    i += 1
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
#Create correlation matrix
train_numeric = train[["cohesion", "syntax", "vocabulary", "phraseology", "grammar", "conventions"]]

sns.heatmap(train_numeric.corr())
```
